# Recommendation system

- i have users and movies data
- user a watched movie 1 ,2,7
- which movie wiil recommend to user a from remaining movies
- compute similarity between usera and user b and usera user b
- use cosine similarity - 0.5,0.6,-0.1
- the one whose cosine similarity is higher then they are similar
- c is similar to a in terms of common movies watched and ratings give
- now see what c has watched
- now recommend a from movies of c
- now see which c has rated as high

```python
movies = pd.read_csv('movies.csv')
ratings = pd.read_csv('ratings.csv')
```

```python
movies.head()
```

```python
ratings.head()
```

```python
# Adding prefix to column data
ratings['userId'] = 'User-'+ratings['userId'].astype(str)
ratings['movieId'] = 'Movie-'+ratings['movieId'].astype(str)
ratings.head()
```

```python
user_item_matrix = ratings.pivot_table(index='userId',columns='movieId',values='rating')
user_item_matrix
```

```python
#AVERAGE NUMBER OF RATINGS PER USER
user_item_matrix.count(axis=1).mean()
```

```python
#user with most number of ratings(jobless user)
user_item_matrix.count(axis=1).sort_values(ascending=False).head(1)
```

```python
#average number of ratings per movie
user_item_matrix.count().sort_values(ascending=False).head(1)
```

```python
from sklearn.metrics.pairwise  import cosine_similarity
#ui_matrix = user_item_matrix.copy()
ui_matrix =user_item_matrix.apply(lambda v: v-v.mean(),axis=1).fillna(0)
ui_matrix.head()
```

```python
sim_matrix = pd.DataFrame(cosine_similarity(ui_matrix),index = ui_matrix.index,columns =ui_matrix.index) 
```

```python
neighs_user100=sim_matrix['User-100'].drop('User-100').sort_values(ascending=False).head(7).index
neighs_user100
```

```python
preds =pd.DataFrame(columns=['mid','predicted_rating'])
for mid,rating in user_item_matrix.loc['User-100'].items():
    if pd.isnull(rating):
        #print(mid)
        mid_ratings =user_item_matrix.loc[neighs_user100,mid]
        avg_mid_rating =mid_ratings.mean()
        preds =preds.append({'mid':mid,'predicted_rating':avg_mid_rating},ignore_index=True)
preds.sort_values('predicted_rating',ascending=False).head(3)
```

```python
preds['movieId'] =preds['mid'].str.replace('Movie-','').astype(int)
preds =pd.merge(preds,movies,left_on='movieId',right_on='movieId')
```

```python
preds.sort_values('predicted_rating',ascending=False).head(3)
```

```python
def ubcf(userid):
    neighs_userid=sim_matrix[userid].drop(userid).sort_values(ascending=False).head(7).index
    preds =pd.DataFrame(columns=['mid','predicted_rating'])
    for mid,rating in user_item_matrix.loc[userid].items():
        if pd.isnull(rating):
        #print(mid)
            mid_ratings =user_item_matrix.loc[neighs_userid,mid]
            avg_mid_rating =mid_ratings.mean()
            preds =preds.append({'mid':mid,'predicted_rating':avg_mid_rating},ignore_index=True)
    #preds.sort_values('predicted_rating',ascending=False).head(3)
    preds['movieId'] =preds['mid'].str.replace('Movie-','').astype(int)
    preds =pd.merge(preds,movies,left_on='movieId',right_on='movieId')
    movies_recom =preds.sort_values('predicted_rating',ascending=False).head(3)['title']
    return movies_recom.values.tolist()
```
