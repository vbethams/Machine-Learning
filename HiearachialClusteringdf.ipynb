```python
import pandas as pd
company = pd.read_csv('company.csv')
```

```python
company.head()
```

```python
company.index =company['Unnamed: 0']
```

```python
company.drop('Unnamed: 0',axis=1,inplace=True)
```

```python
company.head()
```

```python
from sklearn.preprocessing import normalize
company_nor = normalize(company)
from scipy.cluster.hierarchy import dendrogram, linkage
merge = linkage(company_nor,method='complete',metric='euclidean')
merge
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,8))
dendrogram(merge,labels = list(company.index))
plt.show()
```
