```python
import pandas as pd
import numpy as np
mushroom =pd.read_csv('MushRoom.csv')
mushroom.shape
```

```python
mushroom.head()
```

```python
mushroom['class'].value_counts()
```

```python
mush_temp=mushroom.drop('class',axis=1)
```

```python
mush_dummies=pd.get_dummies(mush_temp)
mush_dummies.head()
```

```python
mush_dummies['class']=mushroom['class']
```

```python
from sklearn.model_selection import train_test_split
train,test =train_test_split(mush_dummies,test_size=0.3,random_state=100)
train.head()
```

```python
train_y =train['class']
test_y = test['class']
train_x=train.drop('class',axis=1)
test_x=test.drop('class',axis=1)
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
model.fit(train_x,train_y)
test_pred = model.predict(test_x)
df_pred = pd.DataFrame({'actual':test_y,'predicted':test_pred})
df_pred['pred_status']=df_pred['actual'] ==df_pred['predicted']
df_pred[df_pred['pred_status']==True].shape[0] / df_pred.shape[0] *100
```

# Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
model_rf =RandomForestClassifier(random_state=100)
model_rf.fit(train_x,train_y)

test_pred =model_rf.predict(test_x)
df_pred =pd.DataFrame({'actual':test_y,'predicted':test_pred})
df_pred['pred_status']=df_pred['actual'] == df_pred['predicted']
df_pred[df_pred['pred_status']==True].shape[0]/df_pred.shape[0] *100
```
