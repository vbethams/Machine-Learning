```python
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
```

```python
df = pd.DataFrame({
    'x':[12, 20, 28, 18, 29, 33, 24, 45, 45, 52, 51, 52, 55, 53, 55, 61,64, 69, 71],
    'y':[39, 36, 30, 52, 54, 46, 55, 59, 63, 70, 66, 63, 58, 23, 14, 8, 19, 7, 24]
}) 
```

```python
model = KMeans(n_clusters=3)
model.fit(df)
```

```python
labels = model.predict(df)
labels
```

```python
df['output'] =labels
```

```python
centroids =model.cluster_centers_
centroids
```

```python
import matplotlib.pyplot as plt
fig = plt.figure(figsize=(5,5))
plt.scatter(df['x'],df['y'],c=labels,s=50)
plt.show()
```

```python
plt.scatter(centroids[:,0],centroids[:,1],marker ='*',c='g',s=100)
plt.show()
```

# Silhouette Score

- score is 1 they are good
- if it is -1 it is worst it is overlapping
- if it is 0 both are combined at the boundary

```python
customer = pd.read_csv('Wholesale customers data.csv')
```

```python
customer.head()
```

```python
cust =customer.drop(['Channel','Region'],axis=1)
```

```python
model = KMeans(n_clusters=2)
model.fit(cust)
```

```python
labels = model.labels_
centroids = model.cluster_centers_
```

```python
from sklearn.metrics import silhouette_score
sc=silhouette_score(cust,labels)
sc
```

```python
k=[1,2,3,4,5,6,7,8]
list1=[]
for i in k:
    model = KMeans(n_clusters=i)
    model.fit(cust)
    labels = model.labels_
    #sc=silhouette_score(cust,labels)
    ine = model.inertia_
    list1.append(ine)
    #print('silhoutte score-',i,':',sc)
    
```
